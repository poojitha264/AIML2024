```python
import matplotlib.pyplot as plt
# Generated dataset
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
```

```python
# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(age, height, marker='o', color='blue')
plt.title('Scatter Plot: Age vs Height')
plt.xlabel('Age')
plt.ylabel('Height')
plt.show()
```

```python
# Bar Chart
plt.figure(figsize=(8, 6))
plt.bar(people, weight, color='green')
plt.title('Bar Chart: People vs Weight')
plt.xlabel('People')
plt.ylabel('Weight')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 6))
plt.hist(age, bins=5, color='orange', edgecolor='black')
plt.title('Histogram: Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
